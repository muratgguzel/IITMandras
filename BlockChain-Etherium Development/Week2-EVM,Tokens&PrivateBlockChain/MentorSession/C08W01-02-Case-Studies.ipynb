# <font color='blue'> Quorum </font>

## <font color='blue'> Introduction </font>

Quorum is an open-source permissioned blockchain favoured heavily by enterprises. You would have seen some case studies earlier which are internally using Quorum.

Quorum started as a fork of the geth client/node of Ethereum and hence retains a lot of the basic structure from Ethereum. It was started by JPMorganChase, and development is now run by ConsenSys, a company focused on Ethereum based software.

GoQuorum is the client/node software used in Quorum, similar to geth in Ethereum.

## <font color='blue'> Features </font>


The main goals and accomplishments of Quorum are:
* Performance and Scalability - All the nodes in a private blockchain are known and added by permission, generally. The blockchain is distributed but has some central control. This allows for faster and more deterministic consensus protocols outside of Proof-of-Work and Proof-of-Stake, which are generally used on various public blockchains.

* Private Transactions - The ability to mark transactions private (encrypted), while specifying some public keys as being allowed to see them. This allows transactions or contracts to remain hidden from others but visible to relevant accounts.

* Permission Layer - To create an enterprise level access control layer, Quorum has built a comprehensive permission layer for nodes and accounts, that can model any org or group structure.

* Additional modules on top of Quorum - Quorum provides a lot of pre-created tools and basic modules for common enterprise use cases, thereby making it easier and faster to build on top of Quorum for almost all enterprise use cases.

## <font color='blue'> Consensus Protocols </font>

GoQuorum implements three consensus protocols. We cover them here in a simplified manner, the actual algorithms are, of course, more complex:

* Istanbul Byzantine Fault Tolerant (IBFT) consensus: This is an improvement over the 3-phase distributed commit protocols that we learned earlier. From a fixed list of validators, one proposer is chosen either every block or every round (One round can run for many block confirmations). The default policy is round-robin with every validator becoming the proposer in turn.
The block creation is final without any possible chain forks. Overall, this makes it much faster than the traditional decentralized protocols.

* Raft: This is the fastest among these three with block creation times reaching milliseconds. It has a defined leader, some verifiers/peers/followers, and some learners. Every once in a while, an election can happen among leader/follower nodes to elect a new leader. Learners are similar to a waiting queue of nodes that can be promoted to followers depending on the size requirements.

* Clique (Proof-of-Authority): This is the same as the protocol in Geth. We selected the same while creating the private Ethereum blockchain earlier. In this case, there are pre-defined signers who can sign the block to confirm. Addition and removal of new signers is voted upon by the existing signers. Each signer has a limit to the number of block creations in a period, and they also follow a specific algorithm to limit the number of clashes they can have while trying to sign the same block number.


## <font color='blue'> Private Transactions </font>

Quorum-Design.png



In addition to enhanced RPC apis, Quorum adds a private transaction manager and an enclave (for all crypto-operations) under a Privacy Manager sub-system.

Any transaction can be marked private with access allowed to only specific accounts/nodes. These transactions are stored encrypted on the blockchain.

Tessera (the transaction manager) controls access to encrypted private transactions after validating access. While it manages and orchestrates access, it doesn't have access to the actual encryption/decryption logic or the keys.

The enclave performs all crypto-operations and stores the relevant keys. It essentially behaves like a virtual HSM (Hardware Security Module). This allows easy parallel processing of independent cryptographic operations thereby making the processing even faster.

To accomplish all this, the blockchain structure has been modified from the base Ethereum structure. It stores the public and private transaction data in separate Merkle Patricia Tries, transaction data can be replaced by its encrypted form for creation, storage and hash generation.

## <font color='blue'> Permission Layer </font>

Quorum has made multiple changes for enhanced permissioning schemes. Firstly, it only allows P2P communication with other permissioned nodes. For network permissions, there are two models:

* Basic Network Permissioning: Here, for each node, the nodes that can connect to it, and vice versa, are predefined from the start, based on their enodes listed in a specific json file. This might work when the network is small and specific but can easily get unwieldy


PermissionsModel.png


* Enhanced Permission Model: This manages a very flexible access control model using smart contracts to define the mappings. It includes:

  - Organization and sub-organization units encompassing roles, accounts, and nodes.
  - EOA Accounts
  - Voters - Accounts that can vote on certain actions like block formation
  - Role - Defined roles in orgs
  - Nodes in the network

Admin organization members can manage the overall policies
  


# <font color='blue'> Filecoin and Storj </font>

# <font color='blue'> Introduction </font>


Filecoin and Storj are taking blockchain as a distributed software to the masses for the simplest of use cases - File/Information Storage

Both provide storage support on the blockchain, anyone can save files and information, and anyone can provide their local spare storage! The consumers pay the storage providers in their tokens but all of this is automatically orchestrated based on the storage provided and used. The models that they use are quite different from each other.

# <font color='blue'> Filecoin </font>

Filecoin raised $200+ million in its ICO to provide decentralized storage for all. It's one of the biggest tokens on Ethereum.


Users buy tokens for cash from a currency exchange and spend it for storing their information. Miners (Storage Providers) receive mining rewards for providing storage, and can then sell those tokens to the exchange for cash.

Its based on two crucial proofs:

* **Proof of Replication (PoRep):** Storage miners can prove that they're physically storing a unique copy of the client's data

* **Proof of Spacetime (PoSt)**: Storage miners can prove that they're continuing to dedicate storage to that same data over time.

Filecoin uses a unique CID (Content Identifier) made up of the cryptographic hash and information about which hashing algorithms were used. Since most file data is hierarchical but encrypted, it uses a unique Merkle DAG (Directed Acyclic Graph) to model a tree allowing it to understand and fetch connected data. This allows storage of different files over a lot of nodes but easy search and access to connected files.




# <font color='blue'> Storj </font>

Storj allows 150GB storage and bandwidth free. Over that it charges only \$4 per TB storage and \$7 per TB bandwidth.

Each file that you upload to Storj DCS (Decentralized Cloud Storage) is encrypted, broken up into 80+ redundant chunks, and stored across nodes all over the globe with built-in redundancy. Any 29 pieces are enough to recreate the file. Users can assign access grants to other users on the chain if they want.

There is no single point of failure, a comprehensive security model running on all edges, no system maintenance shutdown, and no chance of decryption outside of your node. It's focused on developers primarily and even has AWS S3 compatible api layer to interact with your files.

There are over 10k nodes in 80+ countries already running edge storage nodes. Nodes chosen for storing chunks of your file are selected by local latency and reputation, with some randomness. If multiple nodes holding your information start going down, the system will auto-repair and select new edge nodes to host your files.
