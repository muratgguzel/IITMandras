# <font color='blue'> Ethereum </font>

As we know that Ethereum was designed and released in 2014-2015. Ethereum had the unique (at that time) plan to have conntinous upgrades which aren't backward compatible as far as the node (EVM) protocol/software is concerned. In this section, we will look at various planned and unplanned updates done over the years on Ethereum.

The releases are generally assigned a name. Here are they in order:

* Frontier - Initial Release (July-2015)
* Iceage (Sep-2015)
* Homestead (March-2016)
* DAO fork - Unplanned (July-2016)
* Tangerine Whistle - Unplanned (Oct-2016)
* Spurious Dragon - Unplanned (Nov-2016)
* Byzantium (Oct-2017)
* Constantinople (Feb-2019)
* Petersburg - Unplanned (Feb-2019)
* Istanbul (Dec-2019)
* Muir Glacier (Jan-2020)
* Berlin (Apr-2021)
* London - To-be-released (Aug-2021)


All of the aforementioned releases were not always planned. Some of the major upgrades in Ethereum blockchain were unplanned. All these changes have made this technology richer as compared to its peers.

Before we start, there is one important mechanism to understand. Ethereum has added a 'difficulty bomb' which will lead to heavy increases in difficulty after a certain block number. 

This was done to ensure that nodes, especially miners, are forced to upgrade to the new version. Specially applicable when the shift from PoW to PoS will happen as that'll adversely affect the miners.

Due to this, once in a while, the difficulty bomb trigger is delayed, until the new changes come in.

Let's start with initial Frontier release

## <font color='blue'> Frontier - Initial Release (July - 2015)  </font>

Frontier is the Ethereum network in its barest form, a beta release essentially: an interface to mine Ether, and a way to upload and execute contracts. 

The main use of Frontier on the launch trajectory is to get mining operations and Ether exchanges running, so the community can get their mining rigs started, and to start to establish a “live” environment where people can test DApps and acquire Ether to upload their own software into Ethereum.

## <font color='blue'> Iceage (September - 2015)  </font>

In this launch, ethereum network created a mechansim to change the difficulty level of the cryptopuzzle at pre-defined levels. 

The Ethereum Ice Age is a difficulty adjustment scheme that was put in place to ensure that everyone has an incentive to move to the updated blockchain once the hard-fork is implemented. 

It's impossible for miners to keep up with the increase in difficulty which would increase block time and make the blockchain freeze, hence the name Ice Age.

## <font color='blue'> Homestead (March - 2016)  </font>

It was the first major update after frontier launch by the the Ethereum organization. Main updates:

* EIP-2 - Increased the cost of contract creation via a transaction, reducing forced incentive to create this way.
* EIP-7 - Addition of DELEGATECALL opcode, which allows for purely computational contracts using sender and value of the caller
* EIP-8 - Allowing forward compatibility in Ethereum clients

## <font color='blue'> DAO Fork - Unplanned (July - 2016)  </font>

This forced change was one of the major event in the Ethereum lifecycle

We have discussed the DAO attack and the resultant fork in detail in the previous notebook.

The event sparked a debate in the crypto-community about whether Ethereum should perform a contentious "hard fork" to reappropriate the affected funds.

It resulted in the network splitting into two blockchains: Ethereum (ETH) with the theft reversed and Ethereum Classic (ETC) which continued on the original chain.

The hard-fork created a rivalry between the two networks. After the hard fork, Ethereum subsequently forked twice more in the 2016 Q4 to deal with other attacks.

## <font color='blue'> Tangerine Whistle - Unplanned (Oct - 2016)  </font>

Tangerine Whistle was a small unplanned fork to counter the denial-of-service attacks happening on Ethereum using transaction spam. Since the cost of some operations was quite low, they were being stuffed in transactions and broadcasted causing computational load in the network.

* EIP-150 - Increase in gas cost for some opcodes like EXTCODESIZE, EXTCODECOPY, BALANCE, SLOAD, CALL, DELEGATECALL, CALLCODE, SELFDESTRUCT

## <font color='blue'> Spurious Dragon - Unplanned (Nov - 2016)  </font>

Spurious Dragon was another unplanned update to counter some attack scenarious that were being noticed.

* EIP-155: We discussed use of nonce in transaction to avoid transaction replay attack on the blockchain. This change solved the problem of replay attacks on different chains - taking a transaction from one testnet and replaying it on the mainnet, for example.
* EIP-160 - EXP cost increase to account for the computational complexity of this operation to avoid it becoming a target to cause network slowdown
* EIP-161 - Allowing removal of empty accounts created in previous attempted DOS attacks, to reduce the state size due to empty accounts.
* EIP-170 - Setting maximum size (24576 bytes) of contract code to avoid fetch of huge data at fixed cost, removing another possible abuse target.

## <font color='blue'> Byzantium (Oct - 2017)  </font>

First half of planned “Metropolis" upgrade, which was the name given to the third planned stage of Ethereum. It contained a wide variety of changes, including 4 new opcodes, 4 precompiles, a reduction of the block reward, and another delay of the Ice Age difficulty bomb. The list includes:
* Change difficulty adjustment to target mean block time including uncles REVERT instruction
* Precompiled contracts for addition and scalar multiplication on the elliptic curve alt_bn128
* Precompiled contracts for optimal ate pairing check on the elliptic curve alt_bn128
* Big integer modular exponentiation
* New opcodes: RETURNDATASIZE and RETURNDATACOPY
* New opcode: STATICCALL
* Metropolis Difficulty Bomb Delay and Block Reward Reduction
* Embedding transaction status code in receipts

## <font color='blue'> Constantinople (Feb - 2019)  </font>

The second half of the Metropolis stage consisted of the Constantinople fork, which was supposed to happen on block 7,080,000, including following 
* Bitwise shifting instructions in EVM
* Skinny CREATE2
* EXTCODEHASH opcode
* Constantinople Difficulty Bomb Delay and Block Reward Adjustment
* Net gas metering for SSTORE without dirty maps

Just before the release, a very specific type of attack known as reentrancy was identfied. The identified attack could have become lethal if the update was released. This issue was result of side effect of one of the updates. 

In order to avoid this side-effect, one of the planned update was removed before the integrations.

## <font color='blue'> Petersburg - Unplanned (Feb-2019)  </font>

As referenced in Constantinopole, this updates removes EIP-1283 added in Constantinopole. It immediately removes the change. 

## <font color='blue'> Istanbul (Dec - 2019)  </font>

Following the Byzantium and Constantinople forks, the name of this fork comes as no surprise to many. The fork includes two new opcodes, a new precompile, and some gas adjustments.
* Add BLAKE2 compression function `F` precompile
* Reduce alt_bn128 precompile gas costs
* ChainID opcode
* Repricing for trie-size-dependent opcodes
* Transaction data gas cost reduction
* Structured Definitions for Net Gas Metering

## <font color='blue'> Muir Glacier (Jan - 2020)  </font>

Muir Glacier only included one update, in order to once again delay the difficulty bomb.

## <font color='blue'> Berlin (Apr - 2021)  </font>

Berlin release has included following updates in the ethereum network:

* Lowers the cost of the ModExp (0x00..05) precompile

* Gas cost increases for state access opcodes, Increases gas cost for SLOAD, *CALL, BALANCE, EXT* and SELFEDESTRUCT when used for the first time in a transaction

* Introduces a new transaction type that is an envelope to enable easier support for multiple transaction types

* Adds a transaction type which contains an access list, a list of addresses and storage keys that the transaction plans to access. This mitigates some of the gas cost increases introduced earlier

## <font color='blue'> London - To-be-released (Aug-2021)  </font>

This is a planned release covering some important changes:

* EIP-1559 - An attempt to reduce the transaction fees for smaller transactions and also reducing the overall Ether supply to induce some deflation
* Another difficulty bomb update


# <font color='blue'> Comparison between Hyperledger, Corda, and Ethereum </font>

Hyperledger, Ethereum, and Corda are some of the popular blockchain platforms at the moment. Let’s take a moment and analyze the comparison of Hyperledger vs Corda vs Ethereum to see how each of these platforms perform.

While the three frameworks showcase the benefits and applications of distributed ledger technologies, they differ a great deal when it comes to vision as well as possible fields of application.

## <font color='blue'> Purpose of platforms </font>

### <font color='blue'> Ethereum </font>

Ethereum seeks to make it easy for developers to come up with smart contracts for performing various tasks. Ethereum blockchain works with generalized protocol allowing everything to run on the network. 

### <font color='blue'> Hyperledger </font> 

Hyperledger, on the other hand, is an open-source collaborative project leveraging blockchain technology to support multiple implementations of components. Hyperledger, on its part, acts as a software whereby people can develop personalized blockchains for tending to various needs.

### <font color='blue'> Corda </font> 
    
Corda’s main aim is to provide people with a platform with common services while ensuring that any service built on top is compatible with network participants.

It also focussed on being business/enterprise compatible.

## <font color='blue'> Peer participation </font>

### <font color='blue'> Ethereum </font>

Ethereum can be both private and public. Anyone can participate in the network at any given time, and also create a private permissioned blockchain environment. 

### <font color='blue'> Hyperledger and Corda </font> 
 

Hyperledger and Corda generally focus on permissioned blockchain networks with a predefined community of participants, with permissioned access to the network.

## <font color='blue'> Consensus Mechanism </font>

### <font color='blue'> Ethereum </font>

Ethereum being a transparent blockchain project essentially means everyone in the project takes part in decision making. For a transaction to be complete in the network, all people in the blockchain must reach a consensus, irrespective of a node being part of the transaction.

Ethereum achieves consensus by means of the Proof-of-Work algorithm whereby all nodes must agree, and all of them must have access to entries recorded. 

### <font color='blue'> Hyperledger </font> 

Hyperledger comes with a different type of consensus whereby nodes are allowed’ to choose between no consensus algorithm and an agreement protocol. In this case, two or more parties can agree and have a significant influence on the outcome. For example, Hyperledger Fabric is using PBFT.

### <font color='blue'> Corda </font> 

Corda comes with two types of consensus:

* Transaction validity: In this case, parties must reach certainty by first checking all associated contract code and come with all the required signatures.

* Transaction Uniqueness: Parties must reach certainty that a transaction in question is a unique consumer of all input stated. This process involves verifying no other transaction consumes any of the agreed states.


## <font color='blue'> Programming Language </font>

### <font color='blue'> Ethereum </font>

Ethereum smart contracts rely on high-level oriented programming languages like Solidity and Vyper.
    
### <font color='blue'> Hyperledger </font> 

Hyperledger relies on ‘chaincode’. The chain code is written in a programming language developed by Google called Golang.
     
### <font color='blue'> Corda </font> 

Corda platform relies on Kotlin, a programming language that is interoperatable with Java and JVM. It can also transcompile to Javascript. By using Kotlin, Corda enjoys a high level of integrations and developer talent since it's being used in a lot of non-blockchain use cases as well.
    
    

# <font color='blue'> Solana </font>

Previously we have covered basics of Solana blockchain tool. In this week, we are going to explore the same tool but this time we will be discussing Leader selection, cluster creation, transaction confirmation and how Proof of History concept is applied here. 

Solana is an open source project implementing a new, high-performance, permissionless blockchain.

A Solana cluster is a set of independently owned computers working together (and sometimes against each other) to verify the output of untrusted, user-submitted programs. 

A Solana cluster can be utilized any time a user wants to preserve an immutable record of events in time or programmatic interpretations of those events.

As long as someone somewhere in the world maintains a copy of the ledger, the output of its programs (which may contain a record of who possesses what) will forever be reproducible, independent of the organization that launched it.

A SOL is the name of Solana's native cryptocurrency, which can be passed to nodes in a Solana cluster in exchange for running an on-chain program or validating its output.

## <font color='blue'> Solana Cluster </font>

A Solana cluster is a set of validators working together to serve client transactions and maintain the integrity of the ledger. Many clusters may coexist. 

When two clusters share a common genesis block, they attempt to converge. Otherwise, they simply ignore the existence of the other. Transactions sent to the wrong one are quietly rejected. 

Before performing any operation on the network, a set of processes need to be completed first. These steps are mentioned below:

## <font color='blue'> Creating a Cluster</font>

Before starting any validator in the network, genesis config should be created. This config refers to public key, mint and bootstrap validator. 

Validator having the bootstrap validator's private key will be responsible for appeding the first entries to ledger. 

Additional validators will then connect with registered validator to become the member of the validator group. 

## <font color='blue'> Joining a Cluster </font>

Solana uses gossip protocol to add more validators/nodes in the cluster. Essentially, it means that regsitering with one validator will lead to propgation to all other nodes in cluster. 

## <font color='blue'> Sending transactions in a Cluster </font>

In this section, tasks will be handled by the nodes based on the allocated role. 

If a node is serving as a validator, it will pass the request to the leader node. 

A node with leader role will then bundle the transactions, timestamp and create an entry and push it into the data plane. 

After that validation will take place and subsequently add it into the ledger. 

## <font color='blue'> Confirmation of transactions </font>

Solana defines confirmation as the duration of time from when the leader timestamps a new entry to the moment when it recognizes a supermajority of ledger votes.

Solona achieves the confirmation by taking the following stamps. 

* Timestamp transactions with a VDF sample and sign the timestamp.

* Split the transactions into batches, send each to separate nodes and have each node share its batch with its peers.

* Repeat the previous step recursively until all nodes have all batches.

The leader therefore timestamps transactions so that validators may lookup the public key of the designated leader. 

The leader then signs the timestamp so that a validator may verify the signature, proving the signer is owner of the designated leader's public key.

Next, transactions are broken into batches so that a node can send transactions to multiple parties without making multiple copies.

For example, if the there are 6 nodes and 60 transactions need to be processed then, Solana is configuered to break that collection of 60 into batches of 10 transactions and send one to each node. 

This allows the leader to put 60 transactions on the wire, not 60 transactions for each node rather a batch of 10 transactions. Each node then shares its batch with its peers. Once the node has collected all 6 batches, it reconstructs the original set of 60 transactions.

This approach will result in efficient usage of network bandwidth. 

## <font color='blue'> Leader rotation </font>

By having only one leader at a time, all validators are able to replay identical copies of the ledger. The drawback of only one leader at a time, however, is that a malicious leader is capable of censoring votes and transactions.

Each validator will choose the leader based on the given algorithm below:

The list of identities of all slot leaders is called a leader schedule. The leader schedule is recomputed locally and periodically. 

It assigns slot leaders for a duration of time called an epoch. The schedule must be computed far in advance of the slots it assigns, such that the ledger state it uses to compute the schedule is finalized. 

That duration is called the leader schedule offset. Solana sets the offset to the duration of slots until the next epoch.

Leader schedule is generated using a predefined seed. The process is as follows:

*Periodically use the PoH tick height (a monotonically increasing counter) to seed a stable pseudo-random algorithm.
* At that height, sample the bank for all the staked accounts with leader identities that have voted within a cluster-configured number of ticks. The sample is called the active set.
* Sort the active set by stake weight.
* Use the random seed to select nodes weighted by stake to create a stake-weighted ordering.
* This ordering becomes valid after a cluster-configured number of ticks.

## <font color='blue'> Proof of History </font>

In any of the distributed system, one of the most difficult problem is agreement of time. 

At Solana another approach named as Proof of History is designed and created blockchain uses this approach.

The idea of Proof of History revolves around a simple question that is "What if it can be proven that a particular message occured sometime before or after the event?"

This concept utilizes the verifiable delay function. This function requires some specific sequential steps to evaluate and produces a unqiue output that is publicly verifiable. 

### <font color='blue'> Upper bound on time </font>

Data can be inserted into the sequence by appending the hash of the data to the previous generated state. The state, input data, and count are all published. Appending the input causes all future output to change unpredictably.

It is still impossible to parallelize, and as long as hash function is pre-image and collision-resistant, it is impossible to create an input that would generate a desired hash in the future, or create an alternative history with the same hashes. 

We can prove that time passed between any two append operations. We can prove that the data was created sometime before it was appended.

### <font color='blue'> Lower bound on time </font>

Inputs into Proof of History can have references back to Proof of History itself. 

The back reference could be inserted as part of a signed message with the users signature, so it cannot be modified without the users private key. 

Because this message contains 0xdeadc0de hash we know it was generated after count 510144806912 was created.


### <font color='blue'> Verification </font>

While the recorded sequence can only be generated on a single CPU core, the output can be verified in parallel.

Each recorded slice can be verified from start to finish on separate cores in 1/(number of cores) time it took to generate. So a modern day GPU with 4000 cores can verify a second in 0.25 milliseconds.
