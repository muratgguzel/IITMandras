## <font color='blue'> Decenralized Application (Dapps) </font>

Dapps are decentralized applications that primarily run on a blockchain. These include any applications like games, NFTs, DeFi applications, exchanges, etc.

The core backend of a dapp will run on a blockchain, made up of a set of smart contracts. The front-end that interfaces with the clients can be offline and written in any technology. There are even decentralized storage solutions like Swarm and IPFS which can be used to store and deliver the front-end.

Other than the examples mentioned in detail, in the sections below, there are various other existing domains where dapps have been created, like games (cryptokitties), prediction markets, freelance marketplaces, etc.

Compared to normal applications, dapps provide some advantages. They are resistant to censorship and attacks. They provide zero downtime and trustless behaviour. They are also verifiable, and can provide privacy and security.

## <font color='blue'> Non-fungible Tokens (NFTs) </font>

We learned earlier about fungible tokens, which are like special currencies driven by standard smart contracts like ERC20, on top of the base cryptocurrency of the blockchain.

Non-fungible tokens, or NFTs, are like one-of-a-kind tokens, analogous to a painting, for example. There is only one Mona Lisa. There might be copies made of it but they are not equivalent to the actual painting.

Similarly, owning an NFT is like owning a unique asset whether directly encoded as a digital asset or a representative ownership of a physical asset. It certifies a unique asset which is not interchangeable with any other.

The most common use case currently is to represent digital assets like important photos, audio files, videos, etc. One can create a digital copy of the same asset, of course, but that doesn't take away the ownership right of that asset.

The NFT market has heated up recently reaching into billion dollars worth of value in 2021. The main use case has been digital art and collectibles as of now. Of course, more serious use cases will become popular like ownership of music and videos which can then be used for direct royalties. We already saw some platforms attempting to do the same.

Ethereum (ERC-721), Flow, and Tezos are the main blockchain platforms being used for NFTs.

## <font color='blue'> Decentralized Finance (DeFi) </font>

Decentralized Finance, or DeFi, is a generic term for financial application on the blockchain. They are primarily focused on moving standard fintech use cases on the blockchain without the need for intermediaries.

A variety of applications can be considered as part of DeFi:
* Decentralized Exchanges (DEX) - Described below in detail
* Lending and loan platforms - Allowing debtors and creditors to interact directly with the ledger and interest rate management done through smart contracts, reducing the need and cost of intermediaries like banks and NBFCs.
* Stablecoins - Normal cryptocurrencies have large variations due to various factors. To do business on the blockchain, there is a need for stable currencies which don't vary drastically on a day-to-day basis. One of the ways this is being accomplished is by creating tokens and keeping them stable by tying their value to a popular fiat currency like USD. Popular stablecoins on Ethereum are Tether (USDT), USD Coin (USDC), and Binance USD (BUSD)
* Trading - Smart contracts on the blockchain allow for a variety of trading scenarios, digital and physical, without the need for trading exchanges, etc.
* Bidding - Various contractual bidding and collectibles sales happen under the guidance of third-parties who take their own cut. In addition, the participants have to trust them as well. Bidding as an application on the blockchain can cut out all of that. Using private blockchains, like Quorum and Hyperledger variants, they can even keep the actual bidding amounts secret while allowing for proof of no fraud.
* Money Legos - Similar to how legos can build bigger things together, using pre-existing and validated contracts for various functionalities together can cover more complex use cases. For example, using oracles, DEX, and lending platforms together to create a comprehensive loan application.
* Insurance - Another big DeFi use case is Insurance. DeFi can force transparent automated claims based on straightforward rules to disrupt the existing insurance market which is opaque and dependent on a lot of hidden rules.

## <font color='blue'> Decentralized Exchange (DEX) </font>

A decentralized exchange, or DEX, allows people to exchange different tokens with each other, similar to currency exchange in fiat currencies like giving INR to receive USD based on the forex rates.

It's a P2P marketplace with exchanges happening directly between peers. The most popular DEXes include Uniswap, Sushiswap, Balancer, etc. They are all built on top of Ethereum.

DEX is a set of smart contracts for facilitating token exchange. Anyone can specify funds in a particular token for exchange. If two sides match interest, then the exchange happens. Prices are determined algorithmically based on the demand and availability in 'liquidity pools'

There is no central authority controlling your assets. There is a huge variety of tokens available to be exchanged. Each transaction will be logged on the blockchain itself making it secure. Unlike centralized exchanges, there is no single attack target to steal a large amount of tokens.

# <font color='blue'> Ethereum 2.0 </font>

The three main problems that Ethereum 2.0 is trying to solve are:

1. Congestion and low TPS: The Ethereum network has become quite popular which has led to heavy network congestion. The current TPS rate is not enough to support all the use cases and achieve the vision of being the world's computing platform. Moreover, the increased demand has led to higher and higher gas prices which lead to exclusion of some applications and use cases which can't afford such high gas prices.

2. Disk space - With more than a billion transactions processed till date, it's extremely hard for full nodes to function. The disk space requirements are high and validation gets very costly.

3. Energy consumption - Similar to Bitcoin, Ethereum still uses Proof-of-Work. At the scale of Ethereum, this consumes a humongous amount of energy every day which is costly and incredibly bad for the environment. The average hash rate of 500k GH/s highlight the number of computing resources, and hence energy, dedicated to just the consensus protocol on Ethereum.

Eth2 is trying to solve these issues within the confines of the trilemma. It wants to scale, without compromising on security and decentralization.

trilemma.png


The process and changes in Eth2 are:

1. Beacon chain: The beacon chain will form the basis of the new expanded network in Ethereum. It provides the base Proof-of-Stake validation and will help coordinate the scaling using shards.

2. Mainnet merge with Beacon chain: Currently, the already shipped beacon chain runs independent of the existing Mainnet. The plan is to merge them and shift the Mainnet to Proof-of-Stake from that point on.

3. Sharing: Finally, the notion of shards will provide the scaling needed on the blockchain. Ethereum's load will get split into 64 shards (new chains). Each transaction can be done on any one of the shards. This would provide a big scaling boost as all the shards can process transactions and smart contract calls in parallel. Moreover, this also makes it easier for the validators as validation cost reduces with small shard sizes and Proof-of-Stake validation. The Beacon chain will coordinate the stakers and assign them to shards. It'll also allow for data exchange between shards so that everyone can share a common Ethereum world state.
