# FleetLINK: Cellular Refrigerated Fleet Management System
- Designed by Bright Wolf for Transport Wireless Technologies
- Transport Wireless Technologies is a leading provider of wireless monitoring devices to the transportation industry. Their solutions provide customers a new level of access to information with products that are easy to understand, install, and service. Driving effciency through lower fuel consumption, labor costs, and preventative maintenance, Transport Wireless Technologies products allow customers to actively monitor & proactively control their yards and fleets.

- Supply Chain Visibility

## Connectivity Protocols	
- Cellular
- MQTT

## Challenge	
1. Create a digital connected transportation solution to retrofit cold chain trailers with real-time tracking and controls.
2. Prevent multi-million dollar losses due to theft or spoilage.
3. Deliver a digital chain-of-custody solution for door to door load monitoring and security.
4. Provide a trusted multi-fleet solution in a single application with granular data and access controls.

## Solution	
1. Real time remote monitoring and history of refrigerated trailer location and status.
2. Full two-way control turns units on or off in single or batch mode, changes set points and modes, and performs diagnostics.
3. Geofence programmable remote locking security ensures trailer doors remain locked beyond controlled areas. 4) Analytics applied to operational data and shared through entire organization.

## Data Collected	
- Fuel Consumption
- Per-Unit Maintenance Costs
- Temperature
- Vehicle Location Tracking
- Labor Costs


## Operational Impact
1. **[Efficiency Improvement - Asset Utilization]**
Better asset utilization through predictive maintenance and data-driven improvement over time.

2. **[Financial Growth - Revenue]**
A differentiating digital solution to reduce load loss and increase value through the entire supply chain

3. **[Efficiency Improvement - Compliance]**
Reliable cold chain compliance for food safety and restricted user access through assigned roles.

## Quantitative Benefit
1. Significant in-yard fuel savings through batch mode commands for pre-cooling trailers rather than a lengthy manual process.
2. Just-in-time route planning due to efficient pre-cool enables flexible delivery routes to take advantage of most recent conditions and market needs.
3. Arrival time notifications ensure load/unload teams are ready and reduce time required to get trucks back on the road.
