# <font color='blue'> 3M Health Information Systems </font>

**3M Health Information Systems (3M HIS)** is a major healthcare solutions provider, and their **3M 360 Encompass System** is used to integrate hospitals' clinical documentation and coding workflows.

A core part of the solution uses natural-language processing to automatically create medical codes, which are critical to both clinicians and the business of healthcare delivery.

3M HIs often faced several raodblocks as part of the development and delivery process:

* Development Time
* Manual Testing
* Manual QA
* Manual Delivery

The organization initially considered building its own private cloud and using solutions from cloud providers, including **Amazon Web Services (AWS)**. After a thorough analysis, 3M HIS found AWS to have the best infrastructure and performance to meet its needs.

3M HIS's applications run on hundreds of **Amazon Elastic Compute Cloud (Amazon EC2)** instances and use **Amazon Simple Storage Service (Amazon S3)**. 

3M HIS also uses **Amazon Relational Database Service (Amazon RDS)**, which helps the organization operate and scale its relational database deployments.

Additionally, 3M HIS leverages:

* The **AWS-managed CodeCommit** source-control service to host private Git repositories,
* **AWS CodeDeploy** to automate deployment, and 
* **AWS CloudFormation** to automate infrastructureâ€”all supporting a DevOps culture, that includes using **AWS CodePipeline** to manage its **Continuous Integration and Continuous Delivery (CI/CD)** deployment workflows.

image.png

They can easily and quickly build application environments and manage applications with AWS CodeDeploy and CodeCommit. Besides, they fully supporting our continuous integration and deployment pipelines by relying on AWS, which boosted their development efficiency.

Deployments have gone from taking six weeks to one per week, and very soon it is expected to touch multiple deployments per day.

As it continues to migrate more of its applications to AWS, 3M HIS expects to create and implement innovative new healthcare software for its customers.

# <font color='blue'> HashedIn: Fully Managed CI/CD Using AWS CodePipeline And EKS </font>

**HashedIn** is a global information technology company that solves complex IT challenges for some of the worldâ€™s largest companies and government organizations.

The **Unav** portal is used by the customer to evaluate architecture and infrastructure best practices for their clients applications, mainly apps hosted on **AWS**.

Unav being a multi-tenant app needed to have a consistent schema for all clients i.e. a common - superset schema for all use cases. For minimizing multiple manual touchpoints, creation of a **CI/CD** pipeline was needed.

HashedIn proposed use of **AWS** **CodeBuild**, **CodeCommit**, **CodePipeline**, and **S3** to create an end-to-end CI/CD pipeline. They also proposed the use of **Infrastructure As Code** - IaC (**Terraform**), to provision infrastructure for an **EKS Cluster** on AWS.

image.png

Three environments were created for **Development**, **QA** and **Production**. The same artifacts would be promoted from Development environment to QA, and finally to Production environment, if approved through multiple stages.

image.png

## <font color='blue'> Commit </font>

Every build triggers with commit on bitbucket using ```pipelines.yml```. AWS CodeCommit replicates the code base, as Code Build was not compatible with bitbucket yet. 

## <font color='blue'> Build </font>

* AWS Codecommit invokes a multi-step build process using fully managed **Continuous Integration**. Once the build is complete, a series of unit test cases and SonarQube analysis are run as per client specified standards. 
* Every build must pass all the checks before it is ready for deployment on the development environment. Once all tests are passed, development build is triggered and checked by development team. 

## <font color='blue'> QA </font>

Post this, the build is promoted manually to QA environment for testing team to check. A series of automated and manual tests are run. Once approved by testing team, QA lead clears the build for promotion to Production. 

## <font color='blue'> Production </font>

* The entire Development and QA infrastructure can be deleted after archival of artifacts on S3. Once production build decision is taken, artifacts stored in S3 are pulled for **Continuous Delivery** on Production with a simple trigger of pipeline.

## <font color='blue'> Advantages Of The AWS CodePipeline Approach </font>

* The CI/CD with Development, QA, and Production environments was prepared on demand, and deleted when not required. This gives significant cost savings to clients.
* Infrastructure created on every trigger of the pipeline ensured that Done-To-Deploy takes a maximum of 20 minutes, instead of at least 3 days taken by any manual process.
